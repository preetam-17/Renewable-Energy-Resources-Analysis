```python
energy_source_usage = re.groupby('Energy_Source').agg({
    'Monthly_Usage_kWh': 'mean',
    'Cost_Savings_USD': 'mean'
}).sort_values('Monthly_Usage_kWh', ascending=False)
print("\nEnergy Source Usage Analysis:")
print(energy_source_usage)
```

```python
plt.figure(figsize=(12, 6))
energy_source_usage.plot(kind='bar', figsize=(14, 8), color=['skyblue', 'salmon'],edgecolor='black')
plt.title('Energy Source Analysis: Monthly Usage and Cost Savings', fontsize=16)
plt.xlabel('Energy Source', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.legend(['Average Monthly Usage (kWh)', 'Average Cost Savings (USD)'], fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create the histogram with KDE
ax = sns.histplot(re['Monthly_Usage_kWh'], bins=30, kde=True, color='#89cff0',edgecolor='purple')

# Annotate the bars with actual frequency values
for patch in ax.patches:
    # Get the x-coordinate (center of the bar) and the height (frequency)
    x = patch.get_x() + patch.get_width() / 2
    y = patch.get_height()
    if y > 0:  # Only annotate bars with a non-zero frequency
        plt.text(x, y, int(y), ha='center', va='bottom', fontsize=10)

# Add title and labels
plt.title('Distribution of Monthly Energy Usage (kWh)', fontsize=16)
plt.xlabel('Monthly Usage (kWh)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Show the plot
plt.show()
```

```python
avg_savings = re.groupby('Energy_Source')['Cost_Savings_USD'].mean().sort_values()
plt.figure(figsize=(10, 6))
colors = ['#FFCFCF','#A888B5','#E82561','#79D7BE','#D39D55']
avg_savings.plot(kind='bar', color=colors,edgecolor='black')
plt.title('Average Cost Savings by Energy Source')
plt.xlabel('Energy Source')
plt.ylabel('Average Cost Savings (USD)') 
plt.xticks(rotation=0, fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Get the counts of energy sources
energy_counts = re['Energy_Source'].value_counts()

# Define a custom color palette
colors = ['#F37D51','#3FCFD4','#EE282A','#6F1B35','#CC6CE7']
# Create the pie chart with the custom colors
plt.pie(energy_counts.values, labels=energy_counts.index, autopct='%1.1f%%', colors=colors)
plt.title('Distribution of Renewable Energy Sources')
plt.axis('equal')  # Ensure the pie chart is a circle
plt.show()
```

```python
cross_tab = pd.crosstab(re['Urban_Rural'], re['Income_Level'])
ax = cross_tab.plot(kind='bar', stacked=False, color=sns.color_palette("Purples", 3), edgecolor='black', figsize=(10, 6))  
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=10, color='black')
plt.title("Stacked Bar Plot of Income Level by Urban/Rural")
plt.xlabel("Urban/Rural")
plt.ylabel("Count")
plt.xticks(rotation=0, fontsize=12)
plt.show()
```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']  # Example color palette
plt.figure(figsize=(10, 6))
sns.countplot(x='Energy_Source', data=re, palette=colors, edgecolor='black')
plt.title('Energy Source Distribution', fontsize=14)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Energy Source', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
numeric_df = re.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
if 'Year' in correlation_matrix.columns:
    correlation_matrix = correlation_matrix.drop('Year', axis=1)
    correlation_matrix = correlation_matrix.drop('Year', axis=0)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.xticks(rotation=0, fontsize=7)
plt.yticks(rotation=0, fontsize=7)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=re, x='Energy_Source', hue='Income_Level', palette='coolwarm', edgecolor='gold',order=re['Energy_Source'].value_counts().index)
plt.title('Energy Source Usage by Income Level', fontsize=16)
plt.xlabel('Energy Source', fontsize=14)
plt.ylabel('Number of Households', fontsize=14)
plt.legend(title='Income Level', fontsize=12, title_fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group re by 'Region' and 'Year' to calculate mean Monthly_Usage_kWh
monthly_usage_by_region_year = re.groupby(['Region', 'Year'])['Monthly_Usage_kWh'].mean().reset_index()

# Plot Monthly Usage by Region and Year
plt.figure(figsize=(12, 8))
sns.lineplot(data=monthly_usage_by_region_year, x='Year', y='Monthly_Usage_kWh', hue='Region', marker='o')
plt.title('Average Monthly Usage (kWh) by Region and Year')
plt.xlabel('Year')
plt.ylabel('Monthly Usage (kWh)')
plt.legend(title='Region')
plt.grid(True)
plt.show()
```
